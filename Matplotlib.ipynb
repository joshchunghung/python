```python
### 增加 x 軸的刻度
import numpy as np
from matplotlib.ticker import MultipleLocator
x=np.linspace(-1,1,50) ### -1~1之間取50個點
y=x**2
plt.plot(x,y,color='red',linewidth=2,linestyle=":")

### xticks
plt.xticks([x for x in np.arange(-1,1.5,0.5)]) ## -1~1之間以0.5取間隔，np.arange算頭不算尾，所以1.5不算,算到1

ax=plt.gca() ### get current axis only for one fig no ax
ax.xaxis.set_minor_locator(MultipleLocator(0.1)) ### 0.1為間隔

ax.set_ylim(1, 0) ### y 往下為正
ax.set_xlim(-1,1)
ax.set_title('Z-comp')
ax.set_xlabel('Amp.')
ax.set_ylabel('Time (s)')

ax.xaxis.set_ticks_position('top')
ax.xaxis.set_label_position('top')

#plt.show()

### save  bbox_inches='tight',pad_inches = 0.05 去白邊
## jpg only for linux
#plt.savefig('test.png',dpi=300,bbox_inches='tight',pad_inches = 0.05)
#plt.close()
```

### 1  fig 與 ax
fig 就是一張圖, ax 則是圖裡的邊框, 若一張圖只有一個邊框，則fig與ax區別不大。

```python
x=np.linspace(-1,1,50) ### -1~1之間取50個點
y=x**2

plt.figure()###figA
plt.plot(x,y,color='red',linewidth=2)


### 到這裡為止,都屬於figA
plt.figure() ###figB
y2=2*x+1
plt.plot(x,y2,color=np.array([0,0,255])/255,linewidth=2)


#plt.savefig('t1.png',dpi=100)
```

```python
### 圖中的邊框就是 ax , 六個邊框六個ax
### constrained_layout=True 要求畫圖時依照各個ax的要求顯示。

fig,ax=plt.subplots(nrows=2, ncols=3, dpi=300,figsize=(8,6),constrained_layout=True)

x=np.linspace(-1,1,50) ### -1~1之間取50個點
y=x**2
ax[1][0].plot(x,y)
for i, ax in enumerate(fig.axes):
    ax.text(0.5, 0.5, f"ax{i}" , va="center", ha="center")
    ax.tick_params(labelbottom=False, labelleft=False)

    
##### fig 分界 上方代碼是同一個fig,下方是另外一個fig
plt.figure() 
y2=2*x+1
plt.plot(x,y2,color=np.array([0,0,255])/255,linewidth=2)
```

```python
fig, ax = plt.subplots(figsize=(12,6))
x = np.linspace(0, 5, 10)
ax.plot(x, x+1, color="blue", linewidth=0.25)
ax.plot(x, x+2, color="blue", linewidth=0.50)
ax.plot(x, x+3, color="blue", linewidth=1.00)
ax.plot(x, x+4, color="blue", linewidth=2.00)

# possible linestype options ‘-‘, ‘--’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color="red", lw=2, linestyle='-')
ax.plot(x, x+6, color="red", lw=2, ls='-.')
ax.plot(x, x+7, color="red", lw=2, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="green", lw=2, ls='--', marker='+')
ax.plot(x, x+10, color="green", lw=2, ls='--', marker='o')
ax.plot(x, x+11, color="green", lw=2, ls='--', marker='s')
ax.plot(x, x+12, color="green", lw=2, ls='--', marker='1',alpha=0.1) ## alpha透明度10%

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='^', markersize=10, markerfacecolor='None')
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
```
