### 1  fig 與 ax
fig 就是一張圖, ax 則是圖裡的邊框, 若一張圖只有一個邊框，則fig與ax區別不大。

```python
x=np.linspace(-1,1,50) ### -1~1之間取50個點
y=x**2

plt.figure()###figA
plt.plot(x,y,color='red',linewidth=2)


### 到這裡為止,都屬於figA
plt.figure() ###figB
y2=2*x+1
plt.plot(x,y2,color=np.array([0,0,255])/255,linewidth=2)


#plt.savefig('t1.png',dpi=100)
```

```python
### 圖中的邊框就是 ax , 六個邊框六個ax
### constrained_layout=True 要求畫圖時依照各個ax的要求顯示。

fig,ax=plt.subplots(nrows=2, ncols=3, dpi=300,figsize=(8,6),constrained_layout=True)

x=np.linspace(-1,1,50) ### -1~1之間取50個點
y=x**2
ax[1][0].plot(x,y)
for i, ax in enumerate(fig.axes):
    ax.text(0.5, 0.5, f"ax{i}" , va="center", ha="center")
    ax.tick_params(labelbottom=False, labelleft=False)

    
##### fig 分界 上方代碼是同一個fig,下方是另外一個fig
plt.figure() 
y2=2*x+1
plt.plot(x,y2,color=np.array([0,0,255])/255,linewidth=2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from glob import glob
from matplotlib.lines import Line2D
import matplotlib.patches as mpatches
from obspy import read
from matplotlib.ticker import MultipleLocator
from matplotlib.legend_handler import HandlerLineCollection, HandlerTuple
```

```python
def getShapeColor(ml,dep):
    """ 
    #https://matplotlib.org/stable/api/markers_api.html
    
    #https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
    
    output: shapeStyle,markersize,depcolor
    
    '#DB453F', '#F2B53A', '#97E838', '#26D2EB', '#7B3AF2'
    
    """
    if ml<4.5:
        shape='^'
        ml2=ml+2
    elif ml<5:
        shape='s'
        ml2=ml+2
    elif ml < 6:
        shape='H'
        ml2=ml+3
    else:
        ml2=ml+4.5
        shape='*'
    
    if dep < 10:
        color='#DB453F'
    elif dep<35:
        color='#F2B53A'
    elif dep<70:
        color='#97E838'
    elif dep<150:
        color='#26D2EB'
    else:
        color='#7B3AF2'
    
    return shape,ml2,color
```

```python
"""
facecolor : 背景顏色 markeredgecolor: legend 框框的顏色

"""

fig,ax=plt.subplots(nrows=1, ncols=2, dpi=300,figsize=(8,6),constrained_layout=True)

colNames=['floor','frq','amp','ml','dep']
for txt in sorted(glob('../test/*.frq_amp.1-2.txt0'),key=lambda name: float(name.split("mag")[1].split("frq")[0].strip("."))):
    df=pd.read_csv(txt, sep=" ",names=colNames)
    marker,markersize,color=getShapeColor(df.ml[0],df.dep[0])
    ax[0].plot(df.amp, df.floor,color=color,linewidth=1,linestyle=":",marker=marker, markersize=markersize, 
    markerfacecolor=color, markeredgewidth=0.5, markeredgecolor='black')

for txt in sorted(glob('../test/*.frq_amp.3.5-5.5.txt0'),key=lambda name: float(name.split("mag")[1].split("frq")[0].strip("."))):
    try:
        df=pd.read_csv(txt, sep=" ",names=colNames)
        marker,markersize,color=getShapeColor(df.ml[0],df.dep[0])
        plt.plot(df.amp, df.floor,color=color,linewidth=1,linestyle=":",marker=marker, markersize=markersize, 
        markerfacecolor=color, markeredgewidth=0.5, markeredgecolor='black')
    except:
        print(txt)
        continue
    
# plt.xticks([x for x in np.arange(0,1.1,0.1)]

# for legend
triangle = Line2D([], [], color='black', marker='^', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=6, label='4.5-')
square = Line2D([], [], color='black', marker='s', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=6, label='5')
hexagon2 = Line2D([], [], color='black', marker='H', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=6, label='6')
star=Line2D([], [], color='black', marker='*', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                        markersize=8, label='6+')

ax[0].set_xlabel('Normalized Amp.')
ax[0].set_ylabel('Floor')
ax[0].set_title('1-2Hz')
ax[0].legend(handles=[triangle,  square, hexagon2,star],loc='upper left',title="ML")
ax[0].set_facecolor('#808080')


dep10 = Line2D([], [], color='#DB453F', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='10')
dep35 = Line2D([], [], color='#F2B53A', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='35')
dep70  = Line2D([], [], color='#97E838', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='70')
dep150 =Line2D([], [], color='#26D2EB', marker='s', linestyle='None', markeredgewidth=0.5,
                        markersize=6, label='150')

dep300 =Line2D([], [], color='#7B3AF2', marker='s', linestyle='None', markeredgewidth=0.5,
                        markersize=6, label='150+')


ax[1].set_xlabel('Normalized Amp.')
ax[1].set_title('3.5-5.5Hz')
ax[1].legend(handles=[dep10,  dep35, dep70,dep150,dep300],loc='upper right',title="EVDP (km)")
ax[1].set_facecolor('#808080')
# fig.savefig('floor_amp.png',dpi=300,bbox_inches='tight',pad_inches = 0.05, facecolor='white')
```

```python
def getShapeColor2(ml,dep):
    """ 
    #https://matplotlib.org/stable/api/markers_api.html
    
    #https://matplotlib.org/stable/tutorials/intermediate/legend_guide.html
    
    output: shapeStyle,markersize,depcolor
    
    '#DB453F', '#F2B53A', '#97E838', '#26D2EB', '#7B3AF2'
    
    """
    if ml<4.5:
        ml2=ml+2
    elif ml<5:
        ml2=ml+4
    elif ml < 5.5:
        ml2=ml+8
    else:
        ml2=ml+8
    
    if dep < 10:
        color='#DB453F'
    elif dep<35:
        color='#F2B53A'
    elif dep<70:
        color='#97E838'
    elif dep<150:
        color='#26D2EB'
    else:
        color='#7B3AF2'
    
    return ml2,color
```

```python
fig,ax=plt.subplots(nrows=1, ncols=1, dpi=300,figsize=(8,6),constrained_layout=True)

colNames=['floor','frq','amp','ml','dep']

for index,txt in enumerate(glob('../test/*.frq_amp.1-2.txt0')):
    df=pd.read_csv(txt, sep=" ",names=colNames)
    df=df[df.floor==9]
    markersize,color=getShapeColor2(float(df.ml),float(df.dep))
    ax.plot(index+1, df.frq,marker='o',color=color, markersize=float(df.ml)+2, linestyle='None',
    markerfacecolor=color, markeredgewidth=0.5, markeredgecolor='black')

for index,txt in enumerate(glob('../test/*.frq_amp.3.5-5.5.txt0')):
    df=pd.read_csv(txt, sep=" ",names=colNames)
    df=df[df.floor==5]
    markersize,color=getShapeColor2(float(df.ml),float(df.dep))
    ax.plot(index+1, df.frq,marker='^',color=color, markersize=float(df.ml)+2, linestyle='None',
    markerfacecolor=color, markeredgewidth=0.5, markeredgecolor='black')
    


dep10 = Line2D([], [], color='#DB453F', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='10')
dep35 = Line2D([], [], color='#F2B53A', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='35')
dep70  = Line2D([], [], color='#97E838', marker='s', linestyle='None', markeredgewidth=0.5,
                          markersize=6, label='70')
dep150 =Line2D([], [], color='#26D2EB', marker='s', linestyle='None', markeredgewidth=0.5,
                        markersize=6, label='150')
dep300 =Line2D([], [], color='#7B3AF2', marker='s', linestyle='None', markeredgewidth=0.5,
                        markersize=6, label='150+')


m4c = Line2D([], [], color='black', marker='o', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=4.5+2)
m5c = Line2D([], [], color='black', marker='o', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=5+4)
m6c = Line2D([], [], color='black', marker='o', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=5.5+8)
m7c=Line2D([], [], color='black', marker='o', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=6+8)



m4t = Line2D([], [], color='black', marker='^', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=4.5+2)
m5t = Line2D([], [], color='black', marker='^', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=5+4)
m6t = Line2D([], [], color='black', marker='^', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=5.5+8)
m7t=Line2D([], [], color='black', marker='^', linestyle='None', markeredgewidth=0.5,markeredgecolor='black',
                          markersize=6+8)


### 兩個圖例畫在同一張圖
legend1 = ax.legend(handles=[dep10,dep35,dep70,dep150,dep300],loc='upper right',ncol=5,title="EVDP (km)")

## 兩種圖案在同一列
ax.legend(handles=[(m4c,m4t),(m5c,m5t),(m6c,m6t),(m7c,m7t)],
          labels=['M4.5', 'M5','M5.5','M6'],
          loc='upper left',
          title="ML",
          ncol=2, #分成兩欄
          handler_map={tuple: HandlerTuple(ndivide=None,pad=1)})# 加這個才會讀到tuple
ax.add_artist(legend1)




ax.set_title(f'Peak Frq.',fontsize=8)
ax.set_xlabel('Event No.')
ax.set_ylabel(f'Frquency (Hz)')
ax.set_xlim([0,30])
ax.set_xticks([x for x in np.arange(0,31,5)])
ax.set_yticks([x for x in np.arange(0,6.5,0.5)])
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.yaxis.set_minor_locator(MultipleLocator(0.1))



# fig.savefig(f'peak_frqAll.png',dpi=300,bbox_inches='tight',pad_inches = 0.05, facecolor='white')
```
